## 创建一个块设备池
```bash
[ceph-wfh@ceph-wfh cluster]$ ceph osd pool create cephfs_rbd 64
pool 'cephfs_rbd' created
[ceph-wfh@ceph-wfh cluster]$ ceph osd pool application enable cephfs_rbd rbd
enabled application 'rbd' on pool 'cephfs_rbd'
[ceph-wfh@ceph-wfh cluster]$ rbd pool init -p cephfs_rbd
```

## 创建块设备镜像
```bash
[ceph-wfh@ceph-wfh cluster]$ rbd create --pool cephfs_rbd rbd_image01 --size 1024
# 查看镜像
[ceph-wfh@ceph-wfh cluster]$ rbd ls cephfs_rbd
rbd_image01

# 更新配置文件
[ceph-wfh@ceph-wfh cluster]$ ceph-deploy --overwrite-conf admin ceph-client1 ceph-osd1 ceph-osd2

# On the ceph-client node, create a block device image.
[ceph-wfh@ceph-client1 ~]$ sudo rbd create rbd_image02 --size 4096 --image-feature layering -m ceph-wfh -k /etc/ceph/ceph.client.admin.keyring -p cephfs_rbd

# 删除块设备镜像
# On the ceph-deploy node
[ceph-wfh@ceph-wfh cluster]$ sudo rbd remove rbd_image01 --pool cephfs_rbd
# Removing image: 100% complete...done.
[ceph-wfh@ceph-wfh cluster]$ sudo rbd remove rbd_image02 --pool cephfs_rbd
# Removing image: 100% complete...done.


# On the ceph-client node, map the image to a block device.
[ceph-wfh@ceph-client1 ~]$ sudo rbd map rbd_image02 --name client.admin -m ceph-wfh -k /etc/ceph/ceph.client.admin.keyring -p cephfs_rbd

Use the block device by creating a file system on the ceph-client node.
```
